## Problem Statement :  Build a recommender system by using cosine simillarties score.


## 1. Import necessary libraries :

```python
import pandas as pd
import numpy as np 
from sklearn.metrics import pairwise_distances

import warnings
warnings.filterwarnings('ignore')
```

## 2. Import Dataset :

```python
book_data = pd.read_csv('book.csv')
book_data
```

## 3. Data Understanding :

```python
book_data.shape
```

```python
book_data.isna().sum()
```

```python
book_data.dtypes
```

## 4. Data Preparation :

```python
book_data.drop(labels='Unnamed: 0',axis=1,inplace=True)
book_data
```

```python
book_data.dtypes
```

```python
book_data.columns = ['user_id','book_title','book_rating']
book_data.head()
```

## 5. Cosine similarities score :

```python
book_data['user_id'].nunique()
```

```python
book_data['book_title'].nunique()
```

```python
book_data['user_id'].unique()
```

```python
cosine_pivot_table = pd.pivot_table(data = book_data,values='book_rating',index='user_id',columns='book_title',).fillna(0) 
cosine_pivot_table.index = book_data['user_id'].unique()
cosine_pivot_table.head(50)
```

```python
cosine_pivot_table.values
```

```python
cosine_ubcf = 1-pairwise_distances(X = cosine_pivot_table.values, metric = 'cosine'  )
cosine_ubcf
```

```python
cosine_ubcf_df = pd.DataFrame(cosine_ubcf)
cosine_ubcf_df.index = book_data['user_id'].unique()
cosine_ubcf_df.columns = book_data['user_id'].unique()
cosine_ubcf_df
```

### Filtering  Data :

```python
cosine_ubcf_df_copy = cosine_ubcf_df.copy()
```

```python
np.fill_diagonal(a = cosine_ubcf_df_copy.to_numpy() ,val = 0 )
```

```python
cosine_ubcf_df_copy
```

```python
cosine_ubcf_df_copy.idxmax()
```

### Top 100 users

```python
first_100_users = cosine_ubcf_df_copy.iloc[:100,:100]
first_100_users
```

```python
first_100_users.idxmax()
```

```python
book_data[(book_data['user_id']==276726) | (book_data['user_id']==277048)]
```

### Inference : 1
**Very less data to suggest any recommendation so try for any other user id.**

```python
book_data[(book_data['user_id']==8) | (book_data['user_id']==9)]
```

## II. Euclidean Distance 

```python
euclidean_pivot_table = pd.pivot_table(data= book_data, values='book_rating',index='user_id',columns='book_title',).fillna(0)
euclidean_pivot_table
```

```python
euclidean_pivot_table.index = book_data['user_id'].unique()
euclidean_pivot_table
```

```python
euclidean_pivot_table.values
```

```python
euclidean_ubcf = pairwise_distances(X = euclidean_pivot_table.values ,metric='euclidean',)
euclidean_ubcf
```

```python
ubcf_euclidean_metric = pd.DataFrame(data =euclidean_ubcf)
ubcf_euclidean_metric.index = book_data['user_id'].unique()
ubcf_euclidean_metric.columns = book_data['user_id'].unique()
ubcf_euclidean_metric.round(2)
```

```python
last_20_users = ubcf_euclidean_metric.iloc[-20:,-20:]
last_20_users
```
